```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Définir le chemin de l'image à charger
chemin_Image = "../VideosOutput/image_rognee_1.jpg"
#  Lire l'image avec OpenCV (en BGR) puis la convertir en RGB pour matplotlib
image = cv2.imread(chemin_Image)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Conversion BGR → RGB pour affichage avec matplotlib
```

```python
# Afficher l'image complète
plt.imshow(image)
plt.show()
```

```python
# --- Extraire la moitié supérieure de l'image ---
# Prendre les lignes de 0 jusqu'à la moitié de la hauteur (shape[0] est la hauteur)
moitie_Haut = image[0: int(image.shape[0] / 2), :]
# Afficher uniquement la moitié supérieure
plt.imshow(moitie_Haut)
plt.show()
```
